```python
# Import necessary libraries
```

```python
from textblob import TextBlob
```

```python
import pandas as pd
```

```python
# Sample tweet data (you can replace this with real tweet text)
```

```python
data = {
    'Tweet': [
        "I love the new phone I bought!",
        "I'm feeling so down today.",
        "It's just an average day.",
        "This product is amazing üòç",
        "Terrible service. Never buying again!",
        "The weather is okay, not bad at all.",
        "Best movie I've seen this year!",
        "Worst experience of my life.",
        "Had a good day at work today.",
        "Nothing special, just a regular morning."
    ]
}
```

```python
# Create DataFrame
```

```python
df = pd.DataFrame(data)
```

```python
# Function to analyze sentiment
def analyze_sentiment(tweet):
    blob = TextBlob(tweet)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
# Apply the function
df['Sentiment'] = df['Tweet'].apply(analyze_sentiment)
```
